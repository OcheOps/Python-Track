```python
# Import the tkinter library for creating the GUI (Graphical User Interface)
from tkinter import *
import random  # Import the random library to let the computer make random choices

# Initialize the main window of the game
root = Tk()
root.geometry('400x400')  # Set the size of the window to 400x400 pixels
root.resizable(0, 0)  # Prevent the window from being resized
root.title('Rock, Paper, Scissors ~ by Hybraid6')  # Set the title of the window
root.config(bg='#41c1ba')  # Set the background color of the window

# Create a heading label for the game title
Label(root, text='Rock, Paper, Scissors Game', font='"Comic Sans MS" 18 bold', bg='#1d4355', fg='#f2f1ec').pack(pady=20)

# Ask the user to input their choice (Rock, Paper, or Scissors)
user_take = StringVar()  # Create a variable to store the user's choice
Label(root, text='Rock, Paper or Scissors?', font='Tahoma 15 bold', bg='#1d4355', fg='#f2f1ec').pack(pady=20)

# Create an entry box where the user can type their choice
Entry(root, font='Tahoma 15', textvariable=user_take, bg='#f2f1ec', fg='#365b6d').pack(pady=10)

# Create a variable to display the result of the game
Result = StringVar()

# Define the function that runs when the user clicks the "PLAY" button
def play():
    user_pick = user_take.get()  # Get the user's choice
    comp_pick = random.choice(['Rock', 'Paper', 'Scissors'])  # Let the computer make a random choice
    if user_pick == comp_pick:
        Result.set(f'Tie, you both selected {comp_pick}')
    elif user_pick == 'Rock':
        if comp_pick == 'Paper':
            Result.set('You lose, computer selected Paper')
        else:
            Result.set('You win, computer selected Scissors')
    elif user_pick == 'Paper':
        if comp_pick == 'Scissors':
            Result.set('You lose, computer selected Scissors')
        else:
            Result.set('You win, computer selected Rock')
    elif user_pick == 'Scissors':
        if comp_pick == 'Rock':
            Result.set('You lose, computer selected Rock')
        else:
            Result.set('You win, computer selected Paper')
    else:
        Result.set('Invalid Choice -- Rock, Paper or Scissors?')  # Handle invalid inputs

# Define the function to reset the game
def Reset():
    Result.set("")  # Clear the result message
    user_take.set("")  # Clear the user's input

# Define the function to exit the game
def Exit():
    root.destroy()  # Close the game window

# Create an entry box to display the result of the game
Entry(root, font='Tahoma 10 bold', textvariable=Result, bg='#f2f1ec', fg='#365b6d', width=50).pack(pady=20)

# Create the "PLAY" button
Button(root, font='"Comic Sans MS" 13 bold', text='PLAY', padx=5, bg='#fe854f', fg='#f2f1ec', command=play).pack(pady=10)

# Create the "RESET" button
Button(root, font='"Comic Sans MS" 13 bold', text='RESET', padx=5, bg='#289dd2', fg='#f2f1ec', command=Reset).pack(side=LEFT, padx=50)

# Create the "EXIT" button
Button(root, font='"Comic Sans MS" 13 bold', text='EXIT', padx=5, bg='#f2f1ec', fg='#365b6d', command=Exit).pack(side=RIGHT, padx=50)

# Run the main loop to keep the window open
root.mainloop()
```
