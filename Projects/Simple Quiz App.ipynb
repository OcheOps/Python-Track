```python
import tkinter as tk  # Import the tkinter library for creating the GUI (Graphical User Interface)
from tkinter import messagebox  # Import messagebox to show pop-up messages

# List of questions and answers
questions = [
    {
        "question": "What is the capital of France?",
        "choices": ["Paris", "London", "Berlin", "Madrid"],
        "answer": "Paris"
    },
    {
        "question": "What is 2 + 2?",
        "choices": ["3", "4", "5", "6"],
        "answer": "4"
    },
    {
        "question": "Who wrote 'Harry Potter'?",
        "choices": ["J.K. Rowling", "J.R.R. Tolkien", "George R.R. Martin", "Stephen King"],
        "answer": "J.K. Rowling"
    }
]

# Initialize variables to keep track of the current question and score
current_question_index = 0
score = 0

# Function to load the current question and choices
def load_question():
    global current_question_index
    question_data = questions[current_question_index]
    question_label.config(text=question_data["question"])  # Set the question text
    for i, choice in enumerate(question_data["choices"]):
        choice_buttons[i].config(text=choice)  # Set the choice button text

# Function to check if the selected answer is correct
def check_answer(selected_choice):
    global current_question_index, score
    question_data = questions[current_question_index]
    if selected_choice == question_data["answer"]:  # If the selected choice is correct
        score += 1  # Increase the score
    current_question_index += 1  # Move to the next question
    if current_question_index < len(questions):
        load_question()  # Load the next question
    else:
        show_result()  # Show the final result

# Function to show the final result
def show_result():
    messagebox.showinfo("Quiz Result", f"Your score: {score}/{len(questions)}")  # Show the score in a message box
    app.quit()  # Close the application

# Create the main window of the game
app = tk.Tk()
app.title("Quiz Game ~ By Hybraid6")  # Set the title of the window
app.config(bg='#41c1ba')  # Set the background color of the window

# Create a label to display the question
question_label = tk.Label(app, text="", font=("Comic Sans MS", 18), bg='#1d4355', fg='#f2f1ec')
question_label.pack(pady=20)

# Create buttons for the answer choices
choice_buttons = []
for i in range(4):
    btn = tk.Button(app, text="", font=("Arial", 14), width=30, bg='#f2f1ec', fg='#365b6d',
                    command=lambda i=i: check_answer(choice_buttons[i].cget("text")))  # When a button is clicked, call the check_answer function
    btn.pack(pady=5)
    choice_buttons.append(btn)  # Store the button in the choice_buttons list

# Load the first question
load_question()

# Run the main loop to keep the window open
app.mainloop()
```
