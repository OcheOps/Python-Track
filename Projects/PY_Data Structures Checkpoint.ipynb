### Question 1
Write a Python program that multiplies all the items in a list.

```python
# Sample list
sample_list = [2, 3, 6]

# Initialize the result to 1 (since multiplying by 0 would always be 0)
result = 1

# Loop through each item in the list and multiply to the result
for item in sample_list:
    result *= item

# Print the final result
print("Result =", result)
```

**Explanation**: 
- We start with a sample list `[2, 3, 6]`.
- We initialize `result` to 1.
- We then loop through each item in the list and multiply `result` by each item.
- Finally, we print the result, which is 36.

### Question 2
Write a Python program to get a list, sorted in increasing order by the last element in each tuple, from a given list of non-empty tuples.

```python
# Sample list of tuples
sample_list = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]

# Sort the list by the last element in each tuple using the sorted function and a lambda function
sorted_list = sorted(sample_list, key=lambda x: x[-1])

# Print the expected result
print("Expected result:", sorted_list)
```

**Explanation**: 
- We start with a sample list of tuples `[(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]`.
- We use the `sorted` function with a lambda function as the key to sort the list by the last element of each tuple.
- We then print the sorted list, which is `[(2, 1), (1, 2), (2, 3), (4, 4), (2, 5)]`.

### Question 3
Write a Python program that combines two dictionaries by adding values for common keys.

```python
# Sample dictionaries
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

# Combine dictionaries by adding values for common keys
combined_dict = d1.copy()  # Start with a copy of the first dictionary
for key, value in d2.items():
    if key in combined_dict:
        combined_dict[key] += value  # Add values for common keys
    else:
        combined_dict[key] = value  # Add new key-value pairs

# Print the expected result
print("Expected result:", combined_dict)
```

**Explanation**:
- We start with two dictionaries `d1` and `d2`.
- We create a copy of `d1` called `combined_dict`.
- We then loop through each key-value pair in `d2`. If the key exists in `combined_dict`, we add the values. If the key does not exist, we add the new key-value pair.
- Finally, we print the combined dictionary, which is `{'a': 400, 'b': 400, 'c': 300, 'd': 400}`.

### Question 4
With a given integral number `n`, write a program to generate a dictionary that contains `(i, i*i)` so that `i` is an integral number between 1 and `n` (both included).

```python
# Given integral number
n = 8

# Generate the dictionary using a dictionary comprehension
result_dict = {i: i*i for i in range(1, n + 1)}

# Print the result
print(result_dict)
```

**Explanation**:
- We start with the given integer `n = 8`.
- We use a dictionary comprehension to create a dictionary where each key `i` (from 1 to `n`) maps to `i*i`.
- Finally, we print the resulting dictionary, which is `{1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64}`.

### Question 5
Write a program to sort a list of tuples by the float element in each tuple.

```python
# Sample list of tuples
sample_list = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]

# Sort the list by the float element in each tuple using the sorted function and a lambda function
sorted_list = sorted(sample_list, key=lambda x: float(x[1]), reverse=True)

# Print the expected result
print("Expected result:", sorted_list)
```

**Explanation**:
- We start with a sample list of tuples `[(item1, 12.20), (item2, 15.10), (item3, 24.5)]`.
- We use the `sorted` function with a lambda function as the key to sort the list by the float value of the second element in each tuple.
- We then print the sorted list, which is `[('item3', '24.5'), ('item2', '15.10'), ('item1', '12.20')]`.

### Question 6
Write a Python program to create a set, iterate over it, and add/remove items.

```python
# Create a set
my_set = {0, 1, 2, 3, 4}

# Print the created set
print("Created set:", my_set)

# Iterate over the set
print("Iterating over the set:")
for item in my_set:
    print(item)

# Add members to the set
my_set.add(5)
print("Set after adding 5:", my_set)

# Remove an item from the set
my_set.remove(3)
print("Set after removing 3:", my_set)
```

**Explanation**:
- We create a set `my_set` with elements `{0, 1, 2, 3, 4}`.
- We print the created set.
- We iterate over the set and print each item.
- We add the number `5` to the set and print the updated set.
- We remove the number `3` from the set and print the updated set.
